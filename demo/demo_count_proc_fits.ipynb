```python
import re,os,glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
# list the folders in the directory
dir_fits = "/wrk-kappa/group/corona/peijin/LC4/fits/t1245_1305/"
sb_num_range = (51,243)
num_fits = 4793
lst_files = sorted(os.listdir(dir_fits))
```

```python
img_num_bucket = []
sb_bucket = []
fits_bucket = []

num_get = lambda x: int(re.search(r'\d+', x).group())

# count the number of fits files in each subband
for idx, sb_num in enumerate(lst_files):
    n_img = len(glob.glob(dir_fits+sb_num+"/*image.fits"))
    sb_bucket.append([num_get(sb_num)]*n_img)
    fits_bucket.append(np.arange(n_img))
```

```python
# get numbers in a string

plt.figure(figsize=(4.5,10))
plt.scatter(np.concatenate(fits_bucket),np.concatenate(sb_bucket),s=0.2)
plt.setp(plt.gca(), xlabel='finished fits', ylabel='SB index', xlim=(0,num_fits), ylim=sb_num_range)
```
